```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
import seaborn as sns
import sklearn
```

```python
df_train = pd.read_csv('train_qWM28Yl.csv')
df_train.head()
```

```python
df_train.info()
```

```python
df_train['is_claim'].value_counts()
```

```python
categorical_features = [cols for cols in df_train if df_train[cols].dtype == 'object']
categorical_features
```

```python
df_train_new1 = pd.get_dummies(data=df_train, columns=[
'area_cluster',
 'segment',
 'model',
 'fuel_type',
 'max_torque',
 'max_power',
 'engine_type',
 'is_esc',
 'is_adjustable_steering',
 'is_tpms',
 'is_parking_sensors',
 'is_parking_camera',
 'rear_brakes_type',
 'transmission_type',
 'steering_type',
 'is_front_fog_lights',
 'is_rear_window_wiper',
 'is_rear_window_washer',
 'is_rear_window_defogger',
 'is_brake_assist',
 'is_power_door_locks',
 'is_central_locking',
 'is_power_steering',
 'is_driver_seat_height_adjustable',
 'is_day_night_rear_view_mirror',
 'is_ecw',
 'is_speed_alert'
], drop_first=True)
df_train_new1
```

```python
df_train_new1.info()
```

```python
df_train_new1.drop(columns = 'policy_id', axis = 1, inplace = True)
df_train_new1.sample(5)
```

```python
from sklearn.utils import resample
#create two different dataframe of majority and minority class 
df_majority = df_train_new1[(df_train_new1['is_claim']==0)] 
df_minority = df_train_new1[(df_train_new1['is_claim']==1)] 
# upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,    # sample with replacement
                                 n_samples= 54844, # to match majority class
                                 random_state=42)  # reproducible results
# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_minority_upsampled, df_majority])
df_upsampled
```

```python
df_upsampled['is_claim'].value_counts()
```

```python
X = df_upsampled.drop(columns = 'is_claim', axis = 1)
y = df_upsampled['is_claim']
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
(X_train, X_validation, y_train, y_validation) = train_test_split(X, y, test_size = 0.1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
rfc=RandomForestClassifier(random_state=42)
```

```python
param_grid = { 
    'bootstrap': [True],
    'n_estimators': [50, 100, 200],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [60, 70, 80, 90, None],
     'min_samples_leaf': [3, 4, 5],
     'min_samples_split': [8, 10, 12],
    'criterion' :['gini', 'entropy']
}
```

```python
CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 5,  n_jobs = -1, verbose = 2)
CV_rfc.fit(X_train, y_train)
```

```python
CV_rfc.best_params_
```

```python
rfc1=RandomForestClassifier(random_state=42, max_features='auto', n_estimators= 200, max_depth=8, criterion='gini')
```

```python
rfc1.fit(X_train, y_train)
```

```python
# performing predictions on the validation dataset
y_pred_validation = rfc.predict(X_validation)
```

```python
# Constructing the confusion matrix.
from sklearn.metrics import confusion_matrix
confusion_matrix(y_validation, y_pred_validation)
```

```python
# To compute the F1 score, simply call the f1_score() function:
from sklearn.metrics import f1_score
f1_score(y_validation, y_pred_validation)
```
